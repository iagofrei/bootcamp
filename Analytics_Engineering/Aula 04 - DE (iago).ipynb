```python
# Criando as variaveis
api_url = "https://api.yelp.com/v3/businesses/search"
api_key = "E75fkz2z6r266cXTaRvTnMSmAuvEq9pqhtsX8oqRxRhP3mJ1wExtiGixZ8wWjNWnX_PgBDtdNoXpm57gWKIxZ0VOZ0emVSC1aOkZFLohRBsiYaxXzHa44-bPZoobZHYx"
headers = {"Authorization":"Bearer {}".format(api_key)}
```

```python
# Fazendo a primeira chamada pra ver o total de itens
params = {
    "term":"bookstore",
    "location":"San Francisco",
    "offset": 0
}

first_result = requests.get(api_url, params = params, headers=headers).json()
# first_result

itens_por_chamada = len(first_result['businesses'])
total = first_result['total']

qtd_de_chamadas = total / itens_por_chamada
arredondando_pra_cima = round(qtd_de_chamadas+0.5)
qtd_de_chamadas_mais_1 = arredondando_pra_cima + 1

resposta = {
    "total": total,
    "itens_por_chamada": itens_por_chamada,
    "qtd_de_chamadas": qtd_de_chamadas,
    "arredondando_pra_cima": arredondando_pra_cima,
    "qtd_de_chamadas_mais_1": qtd_de_chamadas_mais_1
}
print(resposta)


```

```python
add_offset = 50
limite_ = 50
```

```python
offset_ = 0
lista_paginada = []
chamadas = 1
params = {
    "term":"bookstore",
    "location":"San Francisco",
    "offset": offset_,
    "limite": limite_
}


for chamadas in range(1, qtd_de_chamadas_mais_1):
    lista_paginada.extend(
        requests.get(api_url, params=params, headers=headers)
        .json()['businesses']
    )
    chamadas = chamadas + 1
    offset_ = offset_ + add_offset

```

```python
len(lista_paginada)
```

```python
df_paginada = pd.json_normalize(lista_paginada, sep = '_')
df_paginada
```

```python
meta_ = [
    "id",
    "alias",
    "name",
    "image_url",
    "is_closed",
    "url",
    "review_count",
    "rating",
    ["coordinates", "latitude"],
    ["coordinates", "longitude"],
    "transactions",
    "price",
    ["location", "address1"],
    ["location", "address2"],
    ["location", "address3"],
    ["location", "city"],
    ["location", "zip_code"],
    ["location", "country"],
    ["location", "state"],
    ["location", "display_address"],
    "phone",
    "display_phone",
    "distance"
]
paginacao_df = pd.json_normalize(
        lista_paginada,
        record_path='categories',
        sep="_",
        meta=meta_,
        meta_prefix='business_',
        record_prefix='category_',
        errors='ignore'
    )
paginacao_df.shape
```
