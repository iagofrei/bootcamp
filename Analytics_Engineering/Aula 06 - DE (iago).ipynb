# Desafio

* Utilizando a bibliteca faker crie 1000 registro com o método fake.profile()
* Normalize as respostas em uma dataframe
* Crie a coluna idade: "2023" - "ano de nascimento"
* Com base na idade validade quantas pessoas estão dentro da população ativa (entre 18 anos e 65 anos)
* Valide os emails utilizando o regex visto em aula

```python
import pandas as pd
import great_expectations as gx 
```

```python
from datetime import datetime
from faker import Faker

fake = Faker('pt_BR')
```

```python
fake.profile()
```

```python
# profile = []

# for i in range(1000):
#     profile.append(fake.profile())

# df_fake = pd.DataFrame(profile) Não usar DataFrame com um json que é uma lista, use json_normalize
# df_fake

profile = []

for i in range(1000):
    profile.append(fake.profile())

df_fake = pd.json_normalize(profile)
df_fake
```

```python
df_fake['birthdate'] = pd.to_datetime(df_fake['birthdate'], format='%Y-%m-%d')

df_fake
```

```python
ano_atual = datetime.now().year
df_fake['idade'] = ano_atual - pd.DatetimeIndex(df_fake['birthdate']).year

df_fake
```

```python
df_fake.info()
```

```python
df_gx = gx.dataset.PandasDataset(df_fake)
valida_idade = df_gx.expect_column_values_to_be_between(column = 'idade', min_value = 18, max_value = 65)
valida_idade
```

```python
valida_email = df_gx.expect_column_values_to_match_regex(column='mail', regex='^[\w+.]+@\w+\.\w{2,}(?:\.\w{2})?$')
valida_email
```
